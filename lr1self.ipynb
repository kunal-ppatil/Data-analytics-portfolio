```python
import matplotlib.pyplot as plt
from sklearn.liner_model import LinearRegression
```

```python
df=pd.read_csv('../datasets/Salary_Data_1.csv')
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlim([0,12])
plt.ylim([0,130000])
```

```python
X=df[['YearsExperience']]
y=df['Salary']
model = LinearRegression()
model.fit(X,y)
fresher_salary = model.predict([[0]])
y_pred=model.predict([[13]])
plt.scatter(0, fresher_salary, color='green', s=100, label=f'Fresher Salary: {fresher_salary[0]:.0f}')
plt.axhline(y=fresher_salary, color='green', linestyle='--')
y_pred=model.predict(X)
plt.scatter(X,y, color='blue',label='Actual Data')
plt.plot(X,y_pred,color='red',label='Best Fit Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience with Best Fit Line')
plt.legend()
plt.show()
print(f"Line equation: Salary = {model.coef_[0]:.2f} * YearsExperience + {model.intercept_:.2f}")

print(f"Predicted fresher salary (0 years): {fresher_salary[0]:.2f}")

print(f"Predicted  salary (13 years): {experience_salary[13]:.2f}")
```
